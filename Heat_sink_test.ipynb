```python
#### #data recording for test 
####################################
import csv

#each new file must be recorded to a different path 
#must have main path to folder then add in a new ending each time

name = str(input("Please input file name :"))
dirLoc = "C:\\Users\\aaron\\Dropbox\\NewForge\\Work\\Heat_Sink_Europa\\" + name + ".csv" 

header = ['time', 'power', 'airflow', 'heat 1', 'heat 2', 'heat 3','peak']
with open(dirLoc, 'w') as path:
    writer = csv.writer(path)  
    # write the header
    writer.writerow(header)
stop = 0
while (stop != 1):
    air =  0
    power =  140
    h1 =  input('Input h1 reading: ')
    h2 =  input('Input h2 reading: ')
    h3 =  input('Input h3 reading: ')
    with open(dirLoc,  'a', newline='') as path:
        #defines the csv writer to 
        writer = csv.writer(path)  
        #write a new row to the csv file 
        writer.writerow([time.time(),str(power),air,h1,h2,h3])
        path.close()
p = input('peak:')
with open(dirLoc,  'a', newline='') as path:
    #defines the csv writer to 
    writer = csv.writer(path)  
    #write a new row to the csv file 
    writer.writerow(['o','air','h1','h2','h3', str(p)])
    path.close()
print('test done')
```
